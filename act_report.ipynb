## Introduction

The dataset that was analyzed and visualized is the tweet archive of Twitter user @dog_rates, also known as WeRateDogs. WeRateDogs is a Twitter account that rates people's dogs with a humorous comment about the dog. 

The process of analysing and making visualizations of the dataset is made after the process of gathering, assessing and cleaning the datasets or archives provided from different sources. This is the final step, with involves analysing and making visualizations of the cleaned data based in some questions defined.

With the archives provided, it was possible to have three dataset related to this Twitter Account:


* The dataset from `twitter_archive_enhanced.csv`: that contains the WeRateDogs Twitter archive,  that was manually downloaded.


* The dataset from `image_predictions.tsv`: that contains the tweet image predictions, that was manually downloaded, and the url images that the file contains was programmatically downloaded and store in a folder.


* The dataset from `tweet_json.txt`: that contains the additional data provided by Twitter from the Twitter API. By using the Twitter API and its library, the data acquired was stored in the text file, for extraction of additional information.

The questions that were defined for analysis and visualizations are three (but not limited):

1. Which dog stage tend to receive more likes?
2. Has the number of tweet retweets been increasing from year to year?
3. Proportion of tweet sources from the tweets.

## Analysis and Visualization 

For the questions above, it was made Exploratory Data Analysis (EDA) and Visualizations, respectively, and communicated the results after eaxh step.

### Which dog stage tend to receive more likes?

For this question, it is necessary to focus on the dog stages mentioned in the tweets and the it favorite count.

#### EDA

* Histogram plot for Favorites count and Dog Stage distribution:

With median value of 3903.0 and mean value of 8179.204052780396, the plot shows that the distribution is skewed to the right, which means that most of the tweets tend to have less than 8 179 favorites and few of them more than 12 000 favorites.

Most of the tweets in the archive does not mention the dogs stages, which is why the 'None' has the higher distribution number in relation to dogs stages. So for the question, it was worked with the data that contains all the dog stages excepting those that did not contain a dog stage.

#### Visualization

* Bar chart for each dog stage in terms of average favorite count:

It was filtered the data which dog stage was not null and calculated the medium favorite count for each dog stage and represented in the bar plot as follows.

**Puppo** is the most favorited dog stage of all dog stages mentioned in the tweets.

### Has the number of tweet retweets been increasing from year to year?

This question was focused on the retweets of each tweet and the year that the tweets were posted.

#### EDA

* Histogram plot for Retweet count distribution:

The distribution of retweets is skewed to the right, which means that most of the tweets tend to have less than 2 660 retweets and few of them more than 10 000 favorites.

* Scatterplot of Tweet Year in relation to its Retweet count:

Theres is clearly no relationship between the year of the tweet and the number of retweets. 

#### Visualization

* Bar chart for each year in terms of average retweet count:

It is shown that the retweet count increased through the years, besides the lack of relation between the two variables.

### Proportion of tweet sources from the tweets.

For the question, it was gathered the proportion of the number of Twitter source tweets and Vine source tweets. These two are the only tweet sources that the twitter account used for posts.

The results are:
* Twitter source:  2031 
* Vine source:  91

Based in these results, it can be easily shown the proportion of each tweet source:

* Percentage of twitter source tweets:  0.9571159283694628 
* Percentage of vine source tweets:  0.04288407163053723

#### Visualization

* Pie Chart for the tweet sources:

The pie chart reflects the values calculated before. Of all the tweets from the cleaned dataset, 95.71% of the tweets are from the Twitter itself, and just 4.29% are from Vine platform.

## Conclusions

As conclusion of the analysis and visualizations:
    
> Most of the tweets have favorite count less than the mean value of all favorite counts, and most of the tweets does not mention a dog stage too. Besides that, comparing the tweets with dog stages, most of the tweets have the 'Pupper' as dog stage, but the 'Puppo' dog stage, which is one of the minority stages mentioned, is the one that has the most favorite count average of all the stages, followed by the 'Doggo' stage.
